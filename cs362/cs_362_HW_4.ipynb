## Problem 1
Two events $E$ and $F$ are independent if $ \Pr(E \cap F ) = \Pr (E) \Pr(F)$. Now consider tossing a
6-sided fair die. Are the events $E = \{1, 2, 3\}$ and $F = \{4, 5, 6\}$ independent?
The conditional probability for an event $E$ given another event $F$ is the probability that $E$
happens given that $F$ happens, and is denoted by $ \Pr(E|F)$. Please calculate the conditional
probability for events $E = \{1, 2, 3\}$ and $F = \{4, 5, 6\}$ independent.


The value of $\Pr(E|F)$ is the probability $E$ will occur given that $F$ has occured, and if $F$ is given then $E$ cannot occur since they are mutually exclusive events. More explicitly we can say that:
$$ \Pr(E|F) = \dfrac{\Pr(E \cap F)}{\Pr(F)} $$
$$ \Pr(E \cap F) = 0 $$
$$ \Pr(E|F) = 0 $$
$$ \Pr(F|E) = \dfrac{\Pr(F \cap E)}{\Pr(E)} $$
$$ \Pr(F \cap E) = 0 $$
$$ \Pr(F|E) = 0 $$

## Probelm 4
A ship arrives at a port and the 40 sailors on board go ashore for revelry. Later that night,
the 40 sailors return to the ship, all drunk. As a result, each randomly chooses a cabin to sleep
in. (No two sailors share the same cabin.) How many sailors are expected to sleep in their own
cabins?
Hints:

(1) Assume each sailor has his own cabin.\
(2) You may assume that the sailors are coming back sequentially, i.e., one by one and randomly
chooses an unoccupied cabin to sleep if that makes it easier.


Clearly the probability the first sailor getting into his own cabin is $P(s_1) = \frac{1}{40}$ chance. 
For the probability of the second sailor we need two things to be true: 
the first sailor did not lay in sailor 2's bed, $P = \frac{39}{40}$, 
and sailor 2 needs to choose his bed from the remaining given it is still an option, $P=\frac{1}{39}$. 
By that we can say that $P(s_2) = \frac{39}{40} \frac{1}{39} = \frac{1}{40}$. 
The third sailor needs $s_1$ to not lay in his bed, $P = \frac{39}{40}$,
$s_2$ to not lay in his bed, $P=\frac{38}{39}$,
and for himself to lay in the bed that is his, $P=\frac{1}{38}$,
which comes out to $P(s_3) = \frac{39}{40} \frac{38}{39} \frac{1}{38} = \frac{1}{40}$. This pattern will continue and all sailors have a $\frac{1}{40}$ chance in laying in thier own bed. We can say then that if this is true, each sailor has a $\frac{1}{40}$ and there are $40$ sailors so we expect $E(S) = 40 \frac{1}{40} = 1$ sailor to sit in their own bed.

We can do this another way (be it unnessicarily complicated, but what the heck its neat). If we say there is a set of possible final outcomes $ Q = \{0,1,2,...,39,40\}$ where any value of $Q$ is the number of sailors to get in the correct bed. We can then say that the probability of exactly $Q_i$ sailors getting into the correct bed is the probability of $Q_i$ sailors getting into the correct bed, times the probability of $40-Q_i$ sailors getting into the wrong bed, times the number of combinations of that configuration.
$$ P(Q_i) = \left(\dfrac{1}{40} \right)^{Q_i} \left(\dfrac{39}{40} \right)^{40-Q_i} {40 \choose Q_i}$$
We can get the expected value then fromt the equation:
$$ E(Q) = \sum_{Q_i}{Q_iP(Q_i)} = \sum_{Q_i}{Q_i \left(\dfrac{1}{40} \right)^{Q_i} \left(\dfrac{39}{40} \right)^{40-Q_i} {40 \choose Q_i}} $$
Below I calculate this with python and it is 1, shocker!

```python
import math
def prob_4():
    return sum([i*(39)**(40-i)/(40**40)*math.comb(40, i) for i in range(41)])

print(f"The expected value is {prob_4():.3f}.")
```

## Probelm 5
A cereal company is doing a promotion based on the movie *TRANSFORMER*. Each box of
cereal contains one of the $n$ different transformers. Once a customer obtains one of every type
of transformer, he or she can send them in and collect a prize. Assume that the transformer
in each box is put in independently and uniformly at random from the $n$ possibilities and that
customers can’t collaborate with each other to collect the transformers together.

(a) How many boxes of cereal must a customer buy in order to expect collecting every type of
transformer?

(b) If the profit for each box of cereal (including the transformer) is $1, and the company would
like to give away an iPhone 6 (costs about $700), what $n$ should the company choose such
that it will not lose money?

Hints:
- Let $X$ be the random variable of the number of boxes bought until at least every type of
transformer is obtained. Then we need to calculate $E[X]$.
- Let $X_i$ be the number of boxes bought to collect the $i^{th}$ new transformer while you already
have exactly $i − 1$ different transformers, then clearly $X = \sum^{n}_{i=1} X_i$.


(a) If we have 0 transformers, and there are n types for us to get, we can say our probability for getting the first transformer is
$ P(x_0) = \frac{n}{n} = 1 $. The second one is then $P(x_0) = \frac{n-1}{n} = 1 - \frac{1}{n}$. We can see the pattern that 
$$ P(x_i) = \frac{n + i}{n} = 1 - \frac{i}{n}$$
That means that 
$$P(x_{n-1}) = 1 - \frac{n}{n} + \frac{1}{n} = \frac{1}{n}$$

$P(x_i)$ is the chance of getting the $x_i^\text{th}$ transformer per box, the the number of boxes per transformer is 
$E(x_i) = \frac{1}{P(x_i)}$ This implies:
$$ E(x_i) = \dfrac{n}{n-i} $$
The total expectation is the sum of the individiual expectations:
$$ E\left(\sum_{i=0}^{n-1} x_i\right) = \sum_{i=0}^{n-1} {E(x_i)} $$
$$ \Rightarrow E(X) = \sum_{i=0}^{n-1}{\dfrac{n}{n-i}} = n \sum_{i=0}^{n-1} {\dfrac{1}{n-i}} $$
$$ = n \left(\dfrac{1}{n} + \dfrac{1}{n+1} + ... + \dfrac{1}{1} \right) $$
The part in parentheses is a harmonic series of length $n$, denoted $H_n$. We can use the approximation $H_n \approx 0.577 + \log n$ to then get a result that is easier to deal with:
$$ E(X) = n (0.577 + \ln n) $$

(b) If the company does not want to love money, we need the expected value to follow $E(X) \ge 700$.
We can solve $E(X) for $n$ using our approximation:
$$ n (0.577 + \ln n) = 700 \Rightarrow n = 128.79$$
And since we assume we only have whole number of trnasformers that means we need 129 different transformers.
We can also solve it without the approximation by using checka nd guess and that also gives 129.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

f = lambda n: n * (0.577 + math.log(n))
g = lambda n: n*0 + 700

x = np.linspace(1,200, 200)
y = np.array([f(xi) for xi in x])
plt.plot(x, y)
plt.plot(x, g(x))
plt.xticks([129])
plt.grid(True)
plt.title("The intersection of y = 700 and the approx is the solution.")
plt.show()
```

## Problem 7
Suppose you are given an array $A$ of $n$ integers. Suppose that $A$ exhibit a number that appears
more than $n/4$ times. You picked 8 numbers from $A$, where each number is taken uniformly at
random. What is the probability that none of these 8 numbers appear more than $n/4$ times?


If a number appears more than $n/4$ times then all other numbers appear less than $3n/4$ times. That means we can assume that the number of value we will call $X$ is at least 
$|X| = \frac{n}{4}+1$ and $Y$ is less than $\frac{3n}{4}-1$, where $A = X \cup Y$ and $X \cap Y = \emptyset$. We want to know the probability that the number of $x$ values we pick out of 8 will be less than $n/4$. Since we dont know exactly how many values of $x$ we have we will have to give a bound. For some checks we konw that if $n=8$ then we know all the $x$ values will be chosen, and so there is a 0% chance. If $n=32$% then there is a 100% chance since even if all 8 numbers are chosed, $\frac{32}{4} = 8$ so we cannot have more than $n/4$ numbers chosen. The case where all values are $x$ is trivial, since there will always be more than $n/4$ values chosen, until we hit $n=32$ when it is impossible. In the case to maximize not reaching $n/4+1$ values of $x$ being chosen we will assume the minimum number of $x$ values, which is 
$|X| = \lfloor \frac{n}{4} +1 \rfloor$. In this case, to pick 8 values that include more than $n/4$ values of $x$ we have to pick all values of $x$. This implies we can only choose values in $Y$, which is an easy problem. There are $n-x$ values of $y$ and we need to pick $k$ values where 
$k=n-8$, or the number we need to remove to get to 8 remaining (in this view we are removing $k$ values instead of picking 8 but it is the same thing). The probability of removing $k$ values of $Y$ where we start with $y$ values is:
$$ P' = \frac{y}{n} \times \frac{y-1}{n-1} \times \frac{y-2}{n-2} \times  ... \times  \frac{y-k}{n-k} 
= \frac{(n-k)!}{n!} \frac{y!}{(y-k)!}
= \frac{8!}{n!} \frac{y!}{(y-k)!} $$
And this is the probability that there ARE $n/4$ values of $x$ is the final 8, so $P = 1-P'$, and as a reminder this is the case of minimul values of $x$, so this probability is the lower bound. For the range of $8 \le n \le 31$ is:
$$ 0 \le P \le 1 - \frac{8!}{n!} \frac{y!}{(y-k)!} $$
Or in terms of just $n$:
$$ 0 \le P \le 1 - \frac{8!}{n!} \frac{(n-{\lfloor \frac{n}{4} +1 \rfloor})!}{(8 - {\lfloor\frac{n}{4}+1\rfloor})!} $$


```python
import math
import numpy as np
import matplotlib.pyplot as plt
def prob_7(n) -> float:
    x = math.floor(n/4 + 1)
    if x > 8:
        return 1.0
    y = n - x
    k = n-8
    # I use gamma(x+1) instead of x! because it gives me a continuous result for the plot
    p_prime = (math.gamma(8+1)*math.gamma(y+1))/(math.gamma(n+1)*math.gamma(y-k+1))
    p = 1-p_prime
    return p

x = np.linspace(8, 31, 1000)
y = [prob_7(xi) for xi in x]

plt.plot(x, y)
plt.title("The probability of getting less than n/4 of a single number after 8 pulls")
plt.xlabel("n")
plt.xticks(range(8, 32))
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)
plt.ylabel("probability")
plt.show()
```

We can look at what the expected time complexity of quicksort (QS) to be given a randomly ordered set of numbers, and it happens to be $O(n \log n)$. If there was an "adversary" who wanted to order the numbers in a way that will slow down QS they could do so to make the time complexity $O(n^2)$. That means to fight this adversary we can randomize the values of QS and expect $O(n \log n)$. 

Meanwhile the time complexity of insertion sort for a random set of numbers is $O(n^2)$ and if we had an adversary making the worst possible ordering they can only order it so we have $O(n^2)$ anyways. This means that randomizing does not expect fight any unfortunate curcumstances that arize from the ordering of the nunmbers. IN FACT in some cased insertion sort can get down to as low as $O(n)$ if the numbers are already partially sorted, which means randomizing can do more harm than good, and is actually a bad idea.

A concise way to think about this is randomizing will improve an algorithm if the time complexity of a randomly sorted list is closer to the best case than the worst case, which is true in the case of QS and false in the case of insertion sort.
